```python
import matplotlib.pyplot as plt
import numpy as np
import os
import scipy
import umap
```

```python
FD_PATH = 'christina_wav/FD_wav'
MD_PATH = 'christina_wav/MD_wav'
FS = 48000
SONG_DUR_SAMPLES = 70000
```

```python
win = scipy.signal.windows.hann(512)
stft = scipy.signal.ShortTimeFFT(win, 20, FS)

def load_wavs(path):
    wavs = []
    spectros = []
    for filename in sorted(os.listdir(path)):
        print(filename)
        rate, data = scipy.io.wavfile.read(os.path.join(path, filename))
        assert rate == FS
        wav = data[:SONG_DUR_SAMPLES]
        wavs.append(wav)
        spectro = np.log(stft.spectrogram(wav))
        spectro[spectro<-10] = -10
        spectros.append(spectro.flatten())
    return np.vstack(wavs), np.vstack(spectros), len(wavs)

fd_wavs, fd_spectros, num_fd = load_wavs(FD_PATH)
md_wavs, md_spectros, num_md = load_wavs(MD_PATH)
wavs = np.vstack((fd_wavs, md_wavs))
spectros = np.vstack((fd_spectros, md_spectros))
```

```python
def plot_umap(X, titlestr, n_neighbors=10, min_dist=0.1):
    embedding = umap.UMAP(n_neighbors=n_neighbors, n_components=2, min_dist=min_dist).fit_transform(X)
    plt.scatter(embedding[:num_fd, 0], embedding[:num_fd, 1], c='#ff70bb', label='Female-directed')
    plt.scatter(embedding[num_fd:, 0], embedding[num_fd:, 1], c='#467bdc', label='Male-directed', marker='s')
    plt.title(f'UMAP visualization of 54 female-directed songs and 43 male-directed songs,\n{titlestr}, n_neighbors = {n_neighbors}, min_dist = {min_dist}')
    plt.xlabel('UMAP dimension 1')
    plt.ylabel('UMAP dimension 2')
    plt.legend()
    plt.show()
    return embedding
```

```python
_ = plot_umap(wavs, 'waveforms', n_neighbors=5)
_ = plot_umap(wavs, 'waveforms', n_neighbors=10)
_ = plot_umap(wavs, 'waveforms', n_neighbors=20)
_ = plot_umap(wavs, 'waveforms', n_neighbors=40)

_ = plot_umap(spectros, 'spectrograms', n_neighbors=5)
spectro_embedding = plot_umap(spectros, 'spectrograms', n_neighbors=10)
_ = plot_umap(spectros, 'spectrograms', n_neighbors=20)
_ = plot_umap(spectros, 'spectrograms', n_neighbors=40)
```
